# Problem Statement

* Student Attendance Dashboard for Decision Making.
* You will be given Google Sheet with different sheet holding data of different classes.
* Create a dashboard showing summary of what's happening across different classes.
* Which students needs attention for missing session?
* Which students are unsatisfied?
* Which students are very satisfied?
* Column shows attendance.
* R- Column shows rating by the candidate.
* Extra marks for identifying more useful insights.
* Missing values of Type column can be treated as STUDENT.
* This data gets updated daily basis.
* Latest data gets injected in early columns.
* Your code shouldn't need any changes to deal with latest data.
* Data Link - https://docs.google.com/spreadsheets/d/1Y_g6LQI2GLqpKf-fj-rH7D9gfnP2MJzpFA6t0AsBMi8/edit?usp=sharing

```python
import datetime
date_columns = []
def check_date_column(column_name):
    date_string = column_name
    formt = "%d-%m-%y"
    try:
        res = bool(datetime.datetime.strptime(date_string, formt))
    except ValueError:
        res = False
    return res
for i in class_1_data.columns:
    if check_date_column(i):
        date_columns.append(i)
date_columns
```

```python
no_of_classes = len(date_columns)
no_of_classes
```

```python
ranked_columns = []
def check_ranked_column(column_name):
    first_letter = column_name[0]
    date = column_name[2:]
    if first_letter == "R" and check_date_column(date):
        return True
    else:
        return False
for i in class_1_data.columns:
    if check_ranked_column(i):
        ranked_columns.append(i)
ranked_columns
```

```python
highest_rank = class_1_data[ranked_columns].max().max()
highest_rank
```

```python
no_of_ranked_columns = len(ranked_columns)

no_of_ranked_columns
```

```python
type(date_columns)
```

```python
student_roll_nums = list(class_1_data[class_1_data['Type'] == "STUDENT"]['Student Roll Num'])
print(student_roll_nums)
```

```python
student_rank_dict = {}
for i in student_roll_nums:
    sum_of_ranks = list(class_1_data.loc[class_1_data['Student Roll Num'] == i][ranked_columns].sum(axis=1))
    student_rank_dict[i] = (sum_of_ranks[0]/(no_of_classes * 10))*100
print(student_rank_dict)
```

```python
class_1_data.loc[class_1_data['Student Roll Num'] == 2][ranked_columns].sum(axis=1)
```

```python
Avg_ranking = (sum(student_rank_dict.values())/(no_of_students * no_of_classes * 10))*100
Avg_ranking
```

```python
st_att = {}
for i in student_roll_nums:
    attendance = dict(class_1_data.loc[class_1_data['Student Roll Num'] == i][date_columns].sum(axis=0).value_counts())
    st_att[i]=attendance
    if 'Missed' not in attendance:
        st_att[i]['Missed'] = 0
    if 'Attended' not in attendance:
        st_att[i]['Attended'] = 0
for i in student_roll_nums:
    st_att[i]['Student Roll Num'] = i
st_att
```

```python
attendance = class_1_data.loc[class_1_data['Student Roll Num'] == 57][date_columns]
```

```python
for i in st_att:
    if i in student_rank_dict:
        st_att[i]['Student Avg Rank'] = student_rank_dict[i]
        st_att[i]['Class Avg Rank'] = Avg_ranking
st_att
```

```python
for i in st_att:
    if st_att[i]['Student Avg Rank'] < Avg_ranking and st_att[i]['Attended'] <=4:
        st_att[i]["Performance"] = 'Un Satisfied'
    elif st_att[i]['Student Avg Rank'] > Avg_ranking and st_att[i]['Attended'] <=4:
        st_att[i]["Performance"] = 'Satisfied'
    elif st_att[i]['Student Avg Rank'] > Avg_ranking and st_att[i]['Attended'] > 4:
        st_att[i]["Performance"] = 'Very Satisfied'
    elif st_att[i]['Student Avg Rank'] > Avg_ranking and st_att[i]['Attended'] <= 4:
        st_att[i]["Performance"] = 'Satisfied'
    elif st_att[i]['Student Avg Rank'] < Avg_ranking and st_att[i]['Attended'] > 4:
        st_att[i]["Performance"] = 'Un Satisfied'
st_att
```

```python
st_att
new = pd.DataFrame.from_dict(st_att,orient ='index')
new.reset_index(drop=True, inplace=True)
new = new[['Student Roll Num','Missed','Attended','Student Avg Rank','Class Avg Rank','Performance']]
new
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

dict(new['Performance'].value_counts())
```

```python
total_un_satisfied = new['Performance'].value_counts()['Un Satisfied']
total_satisfied = new['Performance'].value_counts()['Satisfied']
total_very_satisfied = new['Performance'].value_counts()['Satisfied']
print(total_un_satisfied,total_satisfied,total_very_satisfied)
```

```python
data = {'Performance': [total_un_satisfied,total_satisfied,total_very_satisfied]}
df = pd.DataFrame(data,columns=['Performance'],index = ['Un satisfied Students','Satisfied Students','Very Satisfied students'])

df.plot.pie(y='Performance',figsize=(7, 7),autopct='%1.1f%%', startangle=180)
plt.show()
```

```python
data = {'Performance': ['Un satisfies','Satisfied','Very Satisfied'],
        'Students': [total_un_satisfied,total_satisfied,total_very_satisfied]}
  
df = pd.DataFrame(data,columns=['Performance','Students'])
df.plot(x ='Performance', y='Students', kind = 'bar')
plt.show()
```
